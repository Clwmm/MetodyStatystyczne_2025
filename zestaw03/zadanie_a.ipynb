```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Generator Boxa–Mullera

```python
def box_muller(n):
    x1 = np.random.rand(n)
    x2 = np.random.rand(n)
    R = np.sqrt(-2 * np.log(x1))
    theta = 2 * np.pi * x2

    y1 = R * np.cos(theta)
    y2 = R * np.sin(theta)

    return np.concatenate([y1, y2])  # 2*n liczb z N(0,1)
```

liczba próbek:

```python
N = 10000
data = box_muller(N)
print(len(data))
```

2. Gęstość normalna f(x)

```python
def normal_pdf(x):
    return (1/np.sqrt(2*np.pi)) * np.exp(-x**2 / 2)
```

zakres do rysowania krzywej

```python
x = np.linspace(-5, 5, 400)
f = normal_pdf(x)
```

 3. Histogram + funkcja analityczna

```python
plt.figure(figsize=(10,6))
plt.hist(data, bins=50, density=True, alpha=0.6, label="Histogram (Box–Muller)")

plt.plot(x, f, 'r', linewidth=2, label="Gęstość analityczna N(0,1)")

plt.xlabel("x")
plt.ylabel("gęstość")
plt.title("Generator N(0,1) metodą Boxa–Mullera + porównanie z PDF")
plt.legend()
plt.grid(True)
plt.show()
```
