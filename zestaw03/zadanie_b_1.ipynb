```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_sums(dist, n_samples=50000, n_sum=10):
    if dist == 'normal':
        data = np.random.randn(n_samples, n_sum)
    elif dist == 'exponential':
        data = np.random.exponential(scale=1.0, size=(n_samples, n_sum))
    elif dist == 'uniform':
        data = np.random.uniform(0, 1, size=(n_samples, n_sum))
    else:
        raise ValueError("Unknown distribution")

    return data.sum(axis=1)
```

Liczba składników sumy

```python
N = 10
```

Generacja

```python
s_normal = generate_sums('normal', n_sum=N)
s_exp = generate_sums('exponential', n_sum=N)
s_unif = generate_sums('uniform', n_sum=N)
```

Rysowanie histogramów

```python
plt.figure(figsize=(12,8))

plt.hist(s_normal, bins=60, density=True, alpha=0.5, label="Suma N(0,1), n="+str(N))
plt.hist(s_exp, bins=60, density=True, alpha=0.5, label="Suma Exponential, n="+str(N))
plt.hist(s_unif, bins=60, density=True, alpha=0.5, label="Suma Uniform, n="+str(N))

plt.legend()
plt.title("Histogramy sum 10 liczb z różnych rozkładów")
plt.grid(True)
plt.show()
```
