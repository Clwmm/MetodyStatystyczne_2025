```python
import numpy as np
import matplotlib.pyplot as plt
```

liczba próbek

```python
n = 50000
```

Generacja

```python
data_normal = np.sort(np.random.randn(n))
data_exp = np.sort(np.random.exponential(size=n))
data_unif = np.sort(np.random.uniform(0, 1, size=n))
```

Różnice między kolejnymi liczbami

```python
diff_normal = np.diff(data_normal)
diff_exp = np.diff(data_exp)
diff_unif = np.diff(data_unif)
```

Rysowanie histogramów

```python
fig, ax = plt.subplots(3, 2, figsize=(12,12))

datasets = [
    ("Normalny", diff_normal),
    ("Wykładniczy", diff_exp),
    ("Jednorodny", diff_unif)
]

for i, (name, diffs) in enumerate(datasets):
    ax[i,0].hist(diffs, bins=80, density=True)
    ax[i,0].set_title(f"Histogram różnic – {name}")
    ax[i,0].set_xlabel("różnica")

    ax[i,1].hist(diffs, bins=80, density=True, log=True)
    ax[i,1].set_title(f"Histogram różnic – {name} (log Y)")
    ax[i,1].set_xlabel("różnica [log]")

plt.tight_layout()
plt.show()
```
